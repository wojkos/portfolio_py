```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
r = requests.get("https://www.otodom.pl/sprzedaz/mieszkanie/warszawa/wola/?search%5Bfilter_float_price%3Ato%5D=500000&search%5Bfilter_float_m%3Ato%5D=100&search%5Bfilter_enum_rooms_num%5D%5B0%5D=3&search%5Bdescription%5D=1&search%5Bdist%5D=0&search%5Bdistrict_id%5D=117&search%5Bsubregion_id%5D=197&search%5Bcity_id%5D=26")
conntent = r.content
soup = BeautifulSoup(conntent, "html.parser")
```

```python
properyty_divs = soup.find_all("article", {"class":"offer-item"})
l = list()
for flat in properyty_divs:
    d = dict()
    d["offer_id"] = flat['class'][1]
    d["offer_name"] = flat.find("span", {"class":"offer-item-title"}).text
    location = flat.select("header p")[0].text.replace("Mieszkanie na sprzedaż: ","")
    location = location.split(', ')
    d["city"] = location[0]
    d["zone"] = location[1]
    d["rooms"] = flat.find("li", {"class":"offer-item-rooms"}).text
    d["area"] = flat.find("li", {"class":"offer-item-area"}).text
    price = flat.find("li", {"class":"offer-item-price"}).text.replace("\n","").replace(" ","")
    d["value"] = re.sub("[a-złA-Z]","", price)
    d["currency"] = re.sub("\d","", price)
    l.append(d)
```
