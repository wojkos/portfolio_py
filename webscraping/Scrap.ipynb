```python
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
url = "https://www.otodom.pl/sprzedaz/mieszkanie/warszawa/wola/?search%5Bfilter_float_price%3Ato%5D=500000&search%5Bfilter_float_m%3Ato%5D=100&search%5Bfilter_enum_rooms_num%5D%5B0%5D=3&search%5Bdescription%5D=1&search%5Bdist%5D=0&search%5Bdistrict_id%5D=117&search%5Bsubregion_id%5D=197&search%5Bcity_id%5D=26"
r = requests.get(url)
scrap_time = int(time.time())
conntent = r.content
soup = BeautifulSoup(conntent, "html.parser")
no_of_pages = int(soup.find("ul", {"class":"pager"}).li.strong.text)
l = list()
```

```python
print(no_of_pages)
```

```python
for page in range(1, no_of_pages + 1):
    r = requests.get(url  + "&page=%s"%page)
    conntent = r.content
    soup = BeautifulSoup(conntent, "html.parser")
    
    properyty_divs = soup.find_all("article", {"class":"offer-item"})
    for flat in properyty_divs:
        d = dict()
        d["offer_id"] = flat['class'][1]
        d["offer_name"] = flat.find("span", {"class":"offer-item-title"}).text
        location = flat.select("header p")[0].text.replace("Mieszkanie na sprzedaż: ","")
        location = location.split(', ')
        d["city"] = location[0]
        d["zone"] = location[1]
        d["rooms"] = flat.find("li", {"class":"offer-item-rooms"}).text
        d["area"] = flat.find("li", {"class":"offer-item-area"}).text
        price = flat.find("li", {"class":"offer-item-price"}).text.replace("\n","").replace(" ","")
        d["value"] = re.sub("[a-złA-Z]","", price)
        d["currency"] = re.sub("\d","", price)
        d["timestamp"] = scrap_time
        l.append(d)
```

```python
import pandas
```

```python
df = pandas.DataFrame(l)
```

```python
def is_csv(file_name):
    return bool(re.match(r".*.csv", file_name))
```

```python
from os import listdir
```

```python
csv_files_list = list(filter(is_csv, listdir("./")))
```

```python
# path =r'C:\DRO\DCL_rawdata_files' # use your path
# allFiles = glob.glob(path + "/*.csv")

list_ = []

for file_ in csv_files_list:
    df = pandas.read_csv(file_,index_col=None, header=0)
    list_.append(df)

frame = pandas.concat(list_, axis = 0)
```
